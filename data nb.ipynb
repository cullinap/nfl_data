```python
import pandas as pd
from bs4 import BeautifulSoup as BS
import requests
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url = 'https://www.pro-football-reference.com/years/2022/fantasy.htm'
res = requests.get(url)
soup = BS(res.content, 'html.parser')
col = soup.findAll('tr')[1]
cols = [i.getText() for i in col.findAll('th')]
rows = soup.find_all('tr')[1:]

player_stats = []
for i in range(len(rows)):
    player_stats.append([col.getText() for col in rows[i].findAll('td')])
```

```python
df = pd.DataFrame(player_stats[1:], columns=cols[1:]); df.head(2)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index().drop('index', axis=1, inplace=True)
```
